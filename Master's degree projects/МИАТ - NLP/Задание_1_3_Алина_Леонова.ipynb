### 1.3 С помощью одного выражения с ``re.sub`` собрать в предложение набор токенов

```python
import re

# преобразую входные данные в строку и избавляюсь от крайних скобок кавычек
# шаблон для замены включает: кавычки и запятые, разграничивающие элементы списка
  # любой символ, кроме букв и цифр, который может быть или не быть (\W)? 
  # и следующие за ним символы (кавычка, запятая, пробел, кавычка) до следующего элемента
# заменяю на первую группу (знак препинания, если он был) (\W)? и пробел
# избавляюсь от лишнего пробела в конце, поскольку в конце предложения должен быть знак конца предложения
make_sentence = lambda c: re.sub(r'[\'\"],\s[\'\"](\W)?([\'\"],\s[\'\"])?', r'\1 ', str(c)[2:-2]) [:-1]


assert make_sentence(["Привет", ",", "как", "дела", "?"]) == 'Привет, как дела?', 'Проверьте регулярное выражение'


# Тестирование
assert make_sentence(["Привет", ",", "как", "дела", "?"]) == 'Привет, как дела?', 'Проверьте регулярное выражение'

text = "Впервые передо мной забрезжил свет потенциальной осмысленности регэкспов: http://docs.python.org/dev/howto/regex.html понятно и доходчиво объясняет их функцию в роли трафарета, — использование групп для выделения конкретных участков текста."
tokens = ['Впервые', 'передо', 'мной', 'забрезжил', 'свет', 'потенциальной', 'осмысленности', 'регэкспов', ':', 'http://docs.python.org/dev/howto/regex.html', 'понятно', 'и', 'доходчиво', 'объясняет', 'их', 'функцию', 'в', 'роли', 'трафарета', ',', '—', 'использование', 'групп', 'для', 'выделения', 'конкретных', 'участков', 'текста', '.']
assert make_sentence(tokens) == text, 'Проверьте регулярное выражение'

# Свои тесты
#assert make_sentence(["Гори", ",", "чтобы", "светить", "."]) == 'Гори, чтобы светить.', 'Не не не'
assert make_sentence(["Smvsk", ":", "Knk", ",", "kmal", ",", "sksd", "(", "jd1", ",", "lm2", ",", "kd3", ")", "+", "-", "ghg", ",", "—", "hrh", "."]) == 'Smvsk: Knk, kmal, sksd (jd1, lm2, kd3) +- ghg, — hrh.', 'Не не не'

print("Тесты пройдены")
```
